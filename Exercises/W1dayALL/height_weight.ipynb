```python
lista_edad_altura = [[29,  1.66], [22, 1.83], [31, 1.80], [27, 1.75], [22,1.82], [35,1.79], [34,1.90], [26,1.78], [26, 1.64], [23, 1.74], [24, 1.85], [28, 1.75], [23, 1.83], [32,1.74], [34, 1.57], [27, 1.58],[38,1.70],[22, 1.79]]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
"""
Clean and normalize data
"""
x = []
y = []
for i in lista_edad_altura:
    x.append(i[0])
    y.append(i[1])
X = np.array(x)
y = np.array(y)
x = X.reshape(-1, 1)
model = LinearRegression()
```

```python
def initial_question():
    """
    Plot de la regresión lineal de los datos
    """
    model.fit(x, y)
    plt.scatter(x,y)
    plt.plot(x, model.predict(x), color="red")
initial_question()

```

```python
"""
Búsqueda del error por cada valor
"""
prediction = model.predict(x)
error = y - prediction
error
```

```python
"""
Usando MSE y RMSE
"""
print(mean_squared_error(y, prediction))
print(np.square(mean_squared_error(y, prediction)))
```

```python
import pandas as pd
def transf_data():
    """
    Transformamos la lista de datos a dataframe para poder sacar la matriz de correlación
    """
    df = pd.DataFrame(lista_edad_altura)
    sns.heatmap(df.corr(), annot=True)
transf_data()
```

Es evidente que no tienen ninguna correlación los datos que tenemos y que poco uso lógico se le puede dar a cualquier predicción que hagamos al respecto ya que el modelo va a aprender cosas incongruentes.
